```python
# Mini Project: File Handling Concept

import os

def create_file():
    filename = input("Enter file name to create: ")
    f = open(filename, "w")
    content = input("Enter content for file: ")
    f.write(content)
    f.close()
    print(f"{filename} created successfully!")

def insert_data():
    filename = input("Enter file name: ")
    f = open(filename, "w")
    text = input("Enter content to insert: ")
    f.write(text)
    f.close()
    print("Content inserted successfully (old content replaced).")

def append_file():
    filename = input("Enter file name: ")
    f = open(filename, "a")
    text = input("Enter content to append: ")
    f.write("\n" + text)
    f.close()
    print("Content appended successfully.")

def read_file():
    filename = input("Enter file name: ")
    try:
        f = open(filename, "r")
        data = f.read()
        print(data)
        f.close()
    except FileNotFoundError:
        print("File not found!")

def replace_content():
    filename = input("Enter file name: ")
    old = input("Enter text to replace: ")
    new = input("Enter new text: ")
    try:
        f = open(filename, "r")
        data = f.read()
        f.close()
        data = data.replace(old, new)
        f = open(filename, "w")
        f.write(data)
        f.close()
        print("Replacement done successfully!")
    except FileNotFoundError:
        print("File not found!")

def delete_file():
    filename = input("Enter file name to delete: ")
    if os.path.exists(filename):
        os.remove(filename)
        print("File deleted successfully!")
    else:
        print("File not found!")

def change_directory():
    path = input("Enter directory path: ")
    try:
        os.chdir(path)
        print("Directory changed successfully!")
    except FileNotFoundError:
        print("Directory not found!")
    except Exception as e:
        print(f"Error: {e}")

def working_directory():
    print("Current Working Directory:", os.getcwd())

def rename_file():
    old_name = input("Enter current file name: ")
    new_name = input("Enter new file name: ")
    try:
        os.rename(old_name, new_name)
        print(f"File renamed from {old_name} to {new_name} successfully!")
    except FileNotFoundError:
        print("File not found!")
    except Exception as e:
        print(f"Error: {e}")


# Menu loop
while True:
    print("\n===== File Handling Menu =====")
    print("1. Create File")
    print("2. Insert into File (overwrite)")
    print("3. Append to File")
    print("4. Read File")
    print("5. Replace Content in File")
    print("6. Delete File")
    print("7. Change Directory")
    print("8. Show Working Directory")
    print("9. Rename File")
    print("10. Exit")
    
    choice = input("Enter your choice (1-10): ")

    if choice == "1":
        create_file()
    elif choice == "2":
        insert_data()
    elif choice == "3":
        append_file()
    elif choice == "4":
        read_file()
    elif choice == "5":
        replace_content()
    elif choice == "6":
        delete_file()
    elif choice == "7":
        change_directory()
    elif choice == "8":
        working_directory()
    elif choice == "9":
        rename_file()
    elif choice == "10":
        print("Exiting... Goodbye!")
        break
    else:
        print("Invalid choice, try again!")
```
